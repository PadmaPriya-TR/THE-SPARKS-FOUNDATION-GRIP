Function to extract text from the Image using pytesseracct

```python
def OCR(image_path):
    img=Image.open(image_path)
    image = cv2.imread(image_path)
    plt.imshow(img)
    string = pytesseract.image_to_string(image)
    print(string)
```

Function to get the bounding box corresponding to each character

```python
def boximage(image_path):
    img=Image.open(image_path)
    image = cv2.imread(image_path)
    imgbox=pytesseract.image_to_boxes(image)
    imgh,imgw,_=image.shape
    for boxes in imgbox.splitlines():
        boxes=boxes.split(" ")
        x,y,w,h=int(boxes[1]),int(boxes[2]),int(boxes[3]),int(boxes[4])
        cv2.rectangle(image, (x,imgh-y),(w, imgh-h) , (0,0,255),3)
    plt.imshow(image)
```
